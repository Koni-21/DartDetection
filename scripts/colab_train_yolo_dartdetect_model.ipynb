```python
import ultralytics
from roboflow import Roboflow
```

```python
rf = Roboflow(api_key="YOUR_API_KEY")
project = rf.workspace("koni").project("dart-segmentation")
dataset = project.version(3).download("yolov8")
```

```python
dataset.location
```

```python
model = ultralytics.YOLO('yolov8n-seg.yaml')
```

```python
dataset_location = "/content/Dart-segmentation-3"
results = model.train(data=f'{dataset_location}/data.yaml', epochs=50, imgsz=[1240,90], rect =True)
```
